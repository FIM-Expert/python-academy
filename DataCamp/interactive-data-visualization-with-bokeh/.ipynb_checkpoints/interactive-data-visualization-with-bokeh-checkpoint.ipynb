```python
import pandas as pd
import numpy as np
from bokeh.io import output_notebook, output_file, show
from bokeh.plotting import figure
```

```python
sp500 = pd.read_excel("sp500_sessions.xlsx")
```

```python
print(sp500.head())
```

```python
plot = figure(plot_width=1000, plot_height=400, x_axis_type="datetime", y_axis_label="S&P 500 Level")

plot.line(x=sp500.Date, y=sp500.Level)

output_notebook()

show(plot)
```
